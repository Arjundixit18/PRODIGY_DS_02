```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Corrected file path using a raw string
file_path = r'C:\Users\dixit\Downloads\titanic.csv'

# Read the CSV file
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print(data.head())
```

```python
data.isnull().sum()
```

```python

data.dropna(subset=["Embarked"], inplace=True)

data = data.assign(Cabin=data["Cabin"].fillna("Unknown"))


data = data.assign(Age=data["Age"].fillna(data["Age"].mean()))


print(data.head())
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python

plt.figure(figsize=(10, 5))
sns.histplot(data["Age"], kde=True, color='skyblue', edgecolor='black')
plt.title("Age Distribution", fontsize=15, fontweight='bold')
plt.xlabel("Age", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(True)
plt.show()
```

```python


plt.figure(figsize=(10, 5))
sns.countplot(data=data, x="Sex", hue="Survived", palette="viridis", edgecolor='black')
plt.title("Survival by Gender", fontsize=15, fontweight='bold')
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Survived", loc="upper left")
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=data, x="Age", y="Fare", hue="Survived", palette="coolwarm", edgecolor='black')
plt.title("Scatter Plot of Age vs Fare", fontsize=15, fontweight='bold')
plt.xlabel("Age", fontsize=12)
plt.ylabel("Fare", fontsize=12)
plt.legend(title="Survived", loc="upper right")
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x="Pclass", hue="Pclass", palette="Set2", edgecolor='black', legend=False)
plt.title("Passenger Class Distribution", fontsize=15, fontweight='bold')
plt.xlabel("Class", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x="Pclass", hue="Survived", palette="Set3", edgecolor='black')
plt.title("Survival Rate by Class", fontsize=15, fontweight='bold')
plt.xlabel("Class", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Survived", loc="upper right")
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=data, x="Survived", y="Fare", hue="Survived", palette="viridis", legend=False)
plt.title("Fare Distribution by Survival", fontsize=15, fontweight='bold')
plt.xlabel("Survived", fontsize=12)
plt.ylabel("Fare", fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()

```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x="Embarked", hue="Embarked", palette="pastel", edgecolor='black', legend=False)
plt.title("Count of Embarked Locations", fontsize=15, fontweight='bold')
plt.xlabel("Embarked", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()
```
